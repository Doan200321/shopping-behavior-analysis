```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Dataset\shopping_behavior_updated_Area.csv')
```

```python
df.info()
```

```python
del df['Unnamed: 0']
```

```python
df.drop(df[df['Area'] == 'East'].index, inplace = True)
```

```python
df.drop(df[df['Area'] == 'Error'].index, inplace = True)
```

```python
df1=df.groupby('Gender').count()
df1['Gender']=['Female','Male']
df1
```

```python
fig = plt.figure(figsize =(5, 3))
plt.pie(df1['Area'], # Giá trị của các nhãn
       labels = df1['Gender'] # Danh sách các nhãn
)

# show plot
plt.title("Biểu đồ tỷ lệ Giới tính theo West")
plt.show()
```

```python
df.drop(df[df['Area'] == 'West'].index, inplace = True)
```

```python
df.drop(df[df['Area']=='Error'].index,inplace=True)
```

```python
df1=df.groupby('Area').sum()
df1['area']=['East','West']
df1
```

```python
fig = plt.figure(figsize =(5, 3))
plt.pie(df1['Area'], # Giá trị của các nhãn
       labels = df1['Gender'] # Danh sách các nhãn
)

# show plot
plt.title("Biểu đồ tỷ lệ Giới tính theo East")
plt.show()
```
